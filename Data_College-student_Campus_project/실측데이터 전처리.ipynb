```python
import glob
import pandas as pd
import numpy as np
output1 = glob.glob('/데이터랩/03_실측데이터/**/*.dat') # 경로설정 중요 , 잘못되면 에러남
```

```python
data_list=[]
for i, v in enumerate(output1):
    i = open(v, 'r')
    data_list.append(i) 
```

```python
def fun(lines):
    a = []
    for i, v in enumerate(lines):
        if "Measurement" in v:
            a.append(i)
    st=[]
    for i, j in zip(a, a[1:]):
        story=lines[i:j]
        st.append(story)
    
    yes=[]
    for j in range(0,119,2):
        for i in range(50):
            try:
                a=st[j][i]
                if " ".join(a.strip().split(" ")[:5]) == "Focal Shift Sag (2,3,4,5) Min":
                    yes.append(st[j])
            except:
                pass
    return yes
```

```python
data=[]
for i in range(794):
        lines=data_list[i].readlines()
        a=fun(lines)
        data.append(a)
```
