```python
import pandas as pd
pd.set_option('display.max_columns', 200)
data_col = ['SP_1','SP_2','SP_3','SP_4','SP_56','ANGLE_1','ANGLE_2','ANGLE_3','ANGLE_4','ANGLE_5','ANGLE_6','L1_S1','L1_S2','L2_S1','L2_S2','L3_S1','L3_S2','L4_S1','L4_S2','L5_S1','L5_S2','L6_S1','L6_S2','F1','F2','F3','F4','F5']
input_data=pd.DataFrame(input_ary,columns=data_col)
#SP THI (0-5)

#LENS ANGLE (6-11)

#SAG S1 S2 순서 (12-23)

#Flange L1~L5/S1 (24-28)
```

```python
np.shape(output_ary)
```

```python
output_data=pd.DataFrame(output_ary)
yield_data=output_data.iloc[:, 64]
```

```python
analysis=input_data.join(yield_data,how='inner')
```

```python
analysis=analysis.rename({64:'yield'}, axis=1)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
%matplotlib inline 

fig, axs = plt.subplots(figsize=(16, 8), ncols=4, nrows=2)
lm_features = ['SP_1','SP_2','SP_3','SP_4','SP_56','ANGLE_1','ANGLE_2','ANGLE_3']
for i, feature in enumerate(lm_features):
    row = int(i/4)
    col = i%4
    #sns.regplot(x=feature, y='yield', data=analysis, ax=axs[row][col])
    sns.boxplot(x=feature, y="yield", data=analysis, ax=axs[row][col], palette="Blues")
```
