```python
import pandas as pd

jido=pd.read_csv(r'C:\Users\stat_17\Desktop\지역좌표.csv', encoding='CP949')

x=jido.iloc[:, 3:4]
y=jido.iloc[:, 4:4+1]

import numpy as np
x=x.values.tolist()
y=y.values.tolist()

x=np.reshape(x, (637))
y=np.reshape(y, (637))
```

```python
vl2_1=pd.read_csv(r'C:\Users\stat_17\Desktop\svt\SHRT_2020061005.csv')
vl2_1.rename(columns = {'2020061005' : 'variable'}, inplace = True)

#좌표 변환 함수#
def abc(a, b, c):
    k= (int(a) +149*(int(b)-1)-1 ) + (37697*c)
    return k

#변환한 행,리스트들 수집#
z = [0, 1, 4, 5, 9 , 14] # 8:강수량 # 6,7 -> PTY, SKY # 10: 최고기온 11: 최저기온 12:SKY2  13:적설량
t =[]
for n in z:
    for i, j in zip(x, y):
        t.append(abc(i, j, n))
        
#수집한 행의 리스트 행 추출#
def ac2_1(a):
    k = vl2_1.iloc[a:a+1,:]
    return k

##
u = []
for i in t:
    u.append(ac2_1(i))
    
    
##
u=np.reshape(u, (3822,23))

##
col2_name1 = vl2_1.columns

##
vl2__1 = pd.DataFrame(u, columns=col2_name1)

##
p=[]
for i in range(6):
    a = vl2__1.iloc[0+637*i:637+637*i,:]
    p.append(a)
    
##
pp=[]
for i in range(6):
    a=p[i].reset_index(drop=True)
    df= jido.join(a, how='left')
    pp.append(df)
    
pp[5]=pp[5].loc[pp[5].variable == 'WAV'].replace(-1.0, 0)


ppp=pd.concat(pp[0:5])
VL2_1=pd.concat([ppp, pp[5]])
```
